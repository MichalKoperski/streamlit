**Zadanie 6: Wczytanie, przygotowanie i wizualizacja danych**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. Wczytanie danych z pliku .npz
data = np.load('Intel_Image.npz', allow_pickle=True)

# Sprawdzenie, jakie są klucze
print("Klucze w słowniku:", list(data.keys()))
```

```python
# 2. wyodrębnienie obrazów i etykiet treningowych i testowych
train_img = data['train_img']
train_lab = data['train_lab']
test_img = data['test_img']
test_lab = data['test_lab']

# 3. Kształt tablic oraz typy i zakres danych
print("Kształt train_img:", train_img.shape)
print("Kształt train_lab:", train_lab.shape)
print("Kształt test_img:", test_img.shape)
print("Kształt test_lab:", test_lab.shape)
```

```python
print("Typ danych obrazów:", train_img.dtype)
print("Zakres wartości pikseli:", train_img.min(), "-", train_img.max())

# Jeśli piksele są w zakresie 0-255, przeskaluję do 0-1
if train_img.max() > 1.0:
    train_img = train_img.astype('float32') / 255.0
    test_img = test_img.astype('float32') / 255.0

# Etykiety do typu całkowitego
train_lab = train_lab.astype(int)
test_lab = test_lab.astype(int)

print("Zakres po skalowaniu:", train_img.min(), "-", train_img.max())
```

```python
categories = ['budynki', 'las', 'lodowiec', 'góry', 'morze', 'ulica']

train_counts = [np.sum(train_lab == i) for i in range(len(categories))]
test_counts = [np.sum(test_lab == i) for i in range(len(categories))]

# 4. Wizualizacja w formie wykresu słupkowego
plt.figure(figsize=(10,5))
plt.bar(categories, train_counts, alpha=0.7, label='Treningowe')
plt.bar(categories, test_counts, alpha=0.7, label='Testowe')
plt.title("Liczba zdjęć w poszczególnych kategoriach")
plt.ylabel("Liczba obrazów")
plt.legend()
plt.show()
```

```python
# 5. Wyświetlenie wybranych obrazów
plt.figure(figsize=(12,6))
for i in range(12):
    idx = np.random.randint(0, len(test_img))
    plt.subplot(3, 4, i+1)
    plt.imshow(test_img[idx])
    plt.title(categories[test_lab[idx]])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

**Zadanie 7: Budowa, trenowanie i ocena modelu sieci neuronowej**

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

np.random.seed(42)
tf.random.set_seed(42)

categories = ['budynki', 'las', 'lodowiec', 'góry', 'morze', 'ulica']
num_classes = len(categories)
input_shape = train_img.shape[1:]

# 1. zbiór
def stratified_cap(X, y, per_class=150): 
    idxs = []
    for c in range(num_classes):
        cls_idx = np.where(y == c)[0]
        if len(cls_idx) == 0: 
            continue
        take = min(per_class, len(cls_idx))
        idxs.append(np.random.choice(cls_idx, size=take, replace=False))
    idxs = np.concatenate(idxs)
    return X[idxs], y[idxs]

Xtr_small, ytr_small = stratified_cap(train_img, train_lab, per_class=120)
Xte_small, yte_small = stratified_cap(test_img,  test_lab,  per_class=120)

print("Używam próbek:", Xtr_small.shape[0], "(train),", Xte_small.shape[0], "(test)")

# 2. 3. Model
model = models.Sequential([
    layers.Conv2D(32, 3, activation='relu', padding='same', input_shape=input_shape),
    layers.MaxPooling2D(2),
    layers.Conv2D(32, 3, activation='relu', padding='same'),
    layers.MaxPooling2D(2),
    layers.GlobalAveragePooling2D(),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name='acc')])

# Pipeline’y tf.data
AUTOTUNE = tf.data.AUTOTUNE
BATCH = 128  # większa paczka = mniej kroków/epok
train_ds = tf.data.Dataset.from_tensor_slices((Xtr_small, ytr_small)).shuffle(2048).batch(BATCH).cache().prefetch(AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((Xte_small, yte_small)).batch(BATCH).cache().prefetch(AUTOTUNE)

# 4. Trening
cb = [EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)]
history = model.fit(train_ds, validation_data=val_ds, epochs=8, callbacks=cb, verbose=1)

# 5. 6. Wykresy krzywych uczenia
hist = history.history
epochs = range(1, len(hist["loss"]) + 1)

plt.figure(figsize=(12,4))

# (a) strata
plt.subplot(1, 2, 1)
plt.plot(epochs, hist["loss"],     marker="o", label="train loss")
plt.plot(epochs, hist["val_loss"], marker="o", label="val loss")
plt.xlabel("Epoka")
plt.ylabel("Strata")
plt.title("Krzywe uczenia – strata")
plt.legend()
plt.grid(True)

# (b) accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, hist["acc"],     marker="o", label="train acc")
plt.plot(epochs, hist["val_acc"], marker="o", label="val acc")
plt.xlabel("Epoka")
plt.ylabel("Dokładność")
plt.title("Krzywe uczenia – accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# 7. Podgląd kilku obrazów ze zbioru testowego 
n_samples = 8
idxs = np.random.choice(len(Xte_small), size=n_samples, replace=False)

X_show = Xte_small[idxs]
y_true = yte_small[idxs]

# predykcje modelu (tu bez tf.data, bezpośrednio na tablicy)
y_proba = model.predict(X_show, verbose=0)
y_pred = np.argmax(y_proba, axis=1)

plt.figure(figsize=(14, 6))

for i in range(n_samples):
    plt.subplot(2, 4, i + 1)
    
    # jeśli obrazy były skalowane do [0,1], to nic nie zmieniamy
    # jeśli są jeszcze w [0,255] i typ np. uint8, też można wyświetlać bez zmiany
    plt.imshow(X_show[i])
    plt.axis("off")
    
    gt_label   = categories[int(y_true[i])]
    pred_label = categories[int(y_pred[i])]
    
    # kolor tytułu: zielony jeśli poprawnie, czerwony jeśli źle
    color = "green" if y_true[i] == y_pred[i] else "red"
    plt.title(f"GT: {gt_label}\nPred: {pred_label}", color=color, fontsize=9)

plt.tight_layout()
plt.show()

# 8. Macierz pomyłek

# 1) Predykcje
train_pred = np.argmax(model.predict(train_ds, verbose=0), axis=1)
test_pred  = np.argmax(model.predict(val_ds,   verbose=0), axis=1)

# Odtworzenie kolejności etykiet
ytr_ordered = np.concatenate([y for _, y in train_ds.unbatch().batch(10)])
yte_ordered = np.concatenate([y for _, y in val_ds.unbatch().batch(10)])

# 2) Dokładność train + test
train_acc = accuracy_score(ytr_ordered, train_pred)
test_acc  = accuracy_score(yte_ordered, test_pred)

print(f"Dokładność modelu:")
print(f" ➤ Train accuracy: {train_acc:.3f}")
print(f" ➤ Test accuracy : {test_acc:.3f}")

# 3) Macierz pomyłek (test)
cm = confusion_matrix(yte_ordered, test_pred, labels=list(range(num_classes)))

plt.figure(figsize=(8, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=categories)
disp.plot(cmap="Blues", xticks_rotation=45)
plt.title("Macierz pomyłek – zbiór testowy")
plt.tight_layout()
plt.show()
```

**9. Wnioski**

Analiza macierzy pomyłek wskazuje, że wytrenowany model osiągnął zróżnicowaną skuteczność pomiędzy klasami. Najlepiej rozpoznawane są klasy las (97/120) oraz lodowiec (96/120) — ich charakterystyczna tekstura oraz odróżniające się kolory ułatwiają klasyfikację.
Najwięcej błędów pojawia się w klasach góry, morze oraz ulica. W szczególności:
* klasa góry jest 65 razy błędnie klasyfikowana jako lodowiec,
* klasa morze aż 74 razy jako lodowiec,
* klasa ulica często mylona jest z budynkami (50 przypadków).
Świadczy to o tendencji modelu do nadmiernego wybierania klasy „lodowiec”, co może sugerować:
* niewystarczającą liczbę cech odróżniających klasy (zbyt prosty model),
* niedobrze zrównoważony zbiór lub wysoki stopień wizualnego podobieństwa (np. dominujące błękity w morzu i lodowcu),
* zbyt małą liczbę próbek (120 na klasę),
* brak augmentacji danych, przez co model nauczył się zbyt małej różnorodności obrazów.
Mimo tych ograniczeń model osiąga rozsądny poziom dokładności na danych testowych, co wskazuje na częściową zdolność do generalizacji. Jednak wyniki pokazują wyraźnie, że sieć ma problemy z klasami o większym zróżnicowaniu wizualnym lub bardziej subtelnych różnicach.
Wnioski końcowe
Aby poprawić jakość klasyfikacji, warto rozważyć:
* zwiększenie liczby próbek treningowych,
* zastosowanie augmentacji (rotacje, jasność, zoom),
* dodanie większej liczby warstw konwolucyjnych i dropout,
* użycie wstępnie wytrenowanych modeli (np. MobileNetV2, ResNet50) jako feature extractora.

**Zadanie 8: Dostrajanie modelu sieci neuronowej**

```python
import numpy as np, tensorflow as tf, itertools, os, random, time
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.callbacks import EarlyStopping

np.random.seed(0); tf.random.set_seed(0)
categories = ['budynki','las','lodowiec','góry','morze','ulica']
NUM_CLASSES = len(categories)
INPUT_SHAPE = train_img.shape[1:]

# Ograniczenie danych
def stratified_cap(X, y, per_class=100):
    idxs=[]
    for c in range(NUM_CLASSES):
        cidx = np.where(y==c)[0]
        take = min(per_class, len(cidx))
        if take>0:
            idxs.append(np.random.choice(cidx, size=take, replace=False))
    idxs = np.concatenate(idxs)
    return X[idxs], y[idxs]

PER_CLASS   = 60          
BATCH       = 128
EPOCHS      = 7
PATIENCE    = 2
SEEDS       = [1,2]        
Xtr, ytr = stratified_cap(train_img, train_lab, PER_CLASS)
Xte, yte = stratified_cap(test_img , test_lab , PER_CLASS)

AUTOTUNE = tf.data.AUTOTUNE
def make_ds(X, y, train=True):
    ds = tf.data.Dataset.from_tensor_slices((X,y))
    if train: ds = ds.shuffle(4096)
    return ds.batch(BATCH).cache().prefetch(AUTOTUNE)

train_ds = make_ds(Xtr, ytr, train=True)
val_ds   = make_ds(Xte, yte, train=False)

# Definicje elementów do strojenia
act_map = {
    "relu": lambda: layers.Activation("relu"),
    "leaky": lambda: layers.LeakyReLU(alpha=0.1),
    "swish": lambda: layers.Activation("swish"),  # tf.nn.silu
}
opt_map = {
    "adam":  lambda: optimizers.Adam(learning_rate=1e-3),
    "rms":   lambda: optimizers.RMSprop(learning_rate=1e-3)
}

# Budowa modelu na bazie konfiguracji
def build_model(cfg):
    Act = act_map[cfg["act"]]

    inputs = layers.Input(shape=INPUT_SHAPE)
    x = inputs
    if cfg["bn_input"]:
        x = layers.BatchNormalization()(x)

    x = layers.Conv2D(cfg["f1"], cfg["k"], padding="same")(x)
    x = Act()(x)
    if cfg["bn_after"]:
        x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(2)(x)

    x = layers.Conv2D(cfg["f2"], cfg["k"], padding="same")(x)
    x = Act()(x)
    if cfg["bn_after"]:
        x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(2)(x)

    if cfg["extra_conv"]:
        x = layers.Conv2D(cfg["f2"], cfg["k"], padding="same")(x)
        x = Act()(x)
        if cfg["bn_after"]:
            x = layers.BatchNormalization()(x)
        x = layers.MaxPooling2D(2)(x)

    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(cfg["dense"], activation=None)(x)
    x = Act()(x)
    if cfg["drop"]>0:
        x = layers.Dropout(cfg["drop"])(x)

    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)
    model = models.Model(inputs, outputs)

    model.compile(
        optimizer=opt_map[cfg["opt"]](),
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name="acc")]
    )
    return model

# 3) Siatka konfiguracji
grid = [
    # BN na wejściu, 2 bloki, średnie filtry, relu vs swish, dropout niski, adam
    {"bn_input":True, "bn_after":False, "extra_conv":False, "f1":32,"f2":32,"k":3,"dense":64,"drop":0.2,"act":"relu","opt":"adam"},
    {"bn_input":True, "bn_after":False, "extra_conv":False, "f1":32,"f2":32,"k":3,"dense":64,"drop":0.2,"act":"swish","opt":"adam"},

    # BN po każdej Conv, dodatkowa warstwa Conv (głębiej), relu, RMSprop
    {"bn_input":False,"bn_after":True , "extra_conv":True , "f1":32,"f2":48,"k":3,"dense":64,"drop":0.3,"act":"relu","opt":"rms"},

    # Więcej filtrów + LeakyReLU + Adam, bez BN wejściowego, BN po Conv
    {"bn_input":False,"bn_after":True , "extra_conv":False, "f1":48,"f2":64,"k":3,"dense":96,"drop":0.4,"act":"leaky","opt":"adam"},

    # Mniejsze filtry + szybka konfiguracja bez BN, swish, RMSprop
    {"bn_input":False,"bn_after":False, "extra_conv":False, "f1":24,"f2":24,"k":3,"dense":64,"drop":0.0,"act":"swish","opt":"rms"},
]

# 4) Pętla treningowa
def fit_and_score(cfg, seed):
    tf.keras.backend.clear_session()
    np.random.seed(seed); tf.random.set_seed(seed); random.seed(seed)

    model = build_model(cfg)
    cb = [EarlyStopping(monitor="val_loss", patience=PATIENCE, restore_best_weights=True, verbose=0)]
    hist = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=cb, verbose=0)

    # Szybka ewaluacja
    ytr_pred = np.argmax(model.predict(train_ds, verbose=0), axis=1)
    yte_pred = np.argmax(model.predict(val_ds,   verbose=0), axis=1)
    ytr_true = np.concatenate([y for _,y in train_ds.unbatch().batch(1024)])
    yte_true = np.concatenate([y for _,y in val_ds.unbatch().batch(1024)])

    train_acc = accuracy_score(ytr_true, ytr_pred)
    test_acc  = accuracy_score(yte_true, yte_pred)
    best_val_loss = np.min(hist.history["val_loss"])
    epochs_run = len(hist.history["val_loss"])
    return model, {"seed":seed, "train_acc":train_acc, "test_acc":test_acc, "val_loss":best_val_loss, "epochs":epochs_run}

results = []
best = {"cfg": None, "seed": None, "model": None, "metrics": None}
best_score = None  # tuple: (test_acc, -val_loss)

start = time.time()
for i, cfg in enumerate(grid, 1):
    for seed in SEEDS:
        model, metrics = fit_and_score(cfg, seed)
        row = {**cfg, **metrics}
        results.append(row)
        # wybieramy po test_acc, w razie remisu niższe val_loss
        score_key = (metrics["test_acc"], -metrics["val_loss"])  # większy lepszy
        if (best_score is None) or (score_key > best_score):
            best_score = score_key
            best = {"cfg": cfg, "seed": seed, "model": model, "metrics": metrics}

        print(f"[{i}/{len(grid)}] seed={seed}  test_acc={metrics['test_acc']:.3f}  "
              f"val_loss={metrics['val_loss']:.3f}  epochs={metrics['epochs']}")
        
elapsed = time.time() - start
print(f"\nGotowe. Przebieg zajal ~{elapsed:.1f}s na tej maszynie.")

# 5) Raport z wyników (posortowany po test_acc desc, potem val_loss asc)
def sort_key(r): return (-r["test_acc"], r["val_loss"])
results_sorted = sorted(results, key=sort_key)
print("\n=== TOP konfiguracje ===")
for r in results_sorted[:10]:
    cfg_part = f"BN_in={r['bn_input']} BN_after={r['bn_after']} extraConv={r['extra_conv']} f={r['f1']}/{r['f2']} k={r['k']} dense={r['dense']} drop={r['drop']} act={r['act']} opt={r['opt']}"
    print(f"seed={r['seed']} | test_acc={r['test_acc']:.3f} | train_acc={r['train_acc']:.3f} | val_loss={r['val_loss']:.3f} | ep={r['epochs']} | {cfg_part}")

# 6) Zapis najlepszego modelu
best_model_path = "best_tuned_cnn.keras"
best["model"].save(best_model_path)
print("\nZapisano najlepszy model:", best_model_path)
print("Najlepsza konfiguracja:", best["cfg"], "seed:", best["seed"])
print("Metryki najlepszego:", best["metrics"])
```
